# Bar plots
## All HCPs

```python
df_plot = pd.DataFrame()

i = 0
for resin in ['cq', 'xq', 'hq']:
    for frac in ['g12', 'eluate']:
        df_plot.at[i, 'resin'] = resin
        df_plot.at[i, 'frac'] = frac
        df_plot.at[i, 'n_hcps'] = len(df_master[df_master[f'{resin}_{frac}']])
        i += 1
```

```python
fig, ax = my_plot.instantiate_fig(x=6, y=6, ylabel='Number of HCPs')
tick_pos = np.arange(0, 3, 1)
ticks = ax.set_xticks(tick_pos, ['Capto Q', 'Poros XQ', 'Poros 50 HQ'], rotation=30, ha='right', rotation_mode='anchor')

width = 1/3
rects_1 = ax.bar(tick_pos - width/2, df_plot[df_plot.frac == 'g12'].n_hcps, width=width, label='End of FT')
rects_2 = ax.bar(tick_pos + width/2, df_plot[df_plot.frac == 'eluate'].n_hcps, width=width, label='Adsorbate')

bar_labels = ax.bar_label(rects_1, fontsize=16)
bar_labels = ax.bar_label(rects_2, fontsize=16)

legend = ax.legend(handlelength=1, frameon=False, fontsize=20, handletextpad=0.8)
ax.set_ylim(0, 1400)

plt.savefig('./images/3_bar_chart.png', dpi=300, bbox_inches='tight')
```

# Split violin plots of biophysical property distributions

```python
resin_name_map = {'cq':'Capto Q', 'xq':'Poros XQ', 'hq':'Poros 50 HQ'}
frac_name_map = {'g12':'End of FT', 'eluate':'Adsorbate'}
```

```python
prop = 'pI_mean'
prop_rename = 'Isoelectric point'

# Prepare a consolidated df_violin
for i, resin in enumerate(['cq', 'xq', 'hq']):
    for j, frac in enumerate(['g12', 'eluate']):
        df_temp = df_master[df_master[f'{resin}_{frac}']].copy()
        df_temp['resin'] = resin_name_map[resin]
        df_temp['frac'] = frac_name_map[frac]
        df_temp.rename(columns={prop:prop_rename}, inplace=True)
        if i + j == 0:
            df_violin = df_temp.copy()
        else:
            df_violin = pd.concat([df_violin, df_temp])

# Make plot
plt.gcf().set_size_inches(5, 6)
ax = sns.violinplot(data=df_violin, x='resin', y=prop_rename, hue='frac', split=True, inner='quartile')

# Modify xtick labels
labels = ax.set_xticklabels(ax.get_xticklabels(),rotation=30, ha="right", rotation_mode='anchor', fontsize=16)
hide = ax.set(xlabel=None)

# Modify legend
handles, labels = ax.get_legend_handles_labels()
order = [1, 0]
ax.legend(handles, labels, frameon=False, fontsize=16, handlelength=1, 
          loc='upper left', bbox_transform=ax.transAxes, bbox_to_anchor=(0.0, 1.02))
ax.set_ylim(2.5, 13.99)

# Modify quartile lines
for l in ax.lines:
    l.set_linestyle(':')
    l.set_linewidth(1)
    l.set_color('white')
for l in ax.lines[1::3]:
    l.set_linestyle('-')
    l.set_linewidth(1)
    l.set_color('white')

# plt.savefig(f'./images/3_split_violin_{prop}.png', dpi=300, bbox_inches='tight')
```

```python
prop = 'pI_mean'
for i, resin in enumerate(['cq', 'xq', 'hq']):
    df_1 = df_master[df_master[f'{resin}_g12']].copy()
    df_2 = df_master[df_master[f'{resin}_eluate']].copy()
    result = ttest_ind(df_1[prop], df_2[prop])
    print(resin, '\t', df_1[prop].mean() < df_2[prop].mean(), '\t', f'{result.pvalue:.3f}')
```
