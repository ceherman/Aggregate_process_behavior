```python
df_aex = pd.read_csv('./generated_tables/swath_master_df_ppm.csv')
df_prev = pd.read_csv('../2022-09-20_sec_proteomics/generated_tables/swath_master_df_ppm.csv')
df_prop = pd.read_csv('./generated_tables/swath_master_df_ppm_with_properties.csv')

df_all = df_prev.merge(df_aex, how='outer')
df_all = df_all.fillna(0)

df_log = df_all.drop(columns=['accn', 'name'])
df_log = df_log.fillna(0)
df_log = np.log10(df_log)
df_log.replace(-1*np.inf, np.nan, inplace=True)
```

```python
resins = ['cq', 'xq', 'hq']
fracs = ['a5', 'b12', 'd9', 'g12', 'eluate']
loadings = ['5% Loading', '28% Loading', '53% Loading', '99% Loading', 'Adsorbate']

resin_names = {'cq':'Capto Q', 'xq':'Poros XQ', 'hq':'Poros 50 HQ'}
frac_names = {'a5':'A5', 'b12':'B12', 'd9':'D9', 'g12':'G12', 'eluate':'Adsorbate'}
percent_loading = {'a5':'5%', 'b12':'28%', 'd9':'53%', 'g12':'99%', 'eluate':'Adsorbate'}
```

```python
min_vals = [df_log['pavin_feed'].min()]
max_vals = [df_log['pavin_feed'].max()]

for j, resin in enumerate(resins):
    for i, frac in enumerate(fracs):
        sample = f'{resin}_{frac}'
        min_vals.append(df_log[sample].min())
        max_vals.append(df_log[sample].max())
        
print(min(min_vals), max(max_vals))
```

```python
fig, ax = plt.subplots(3, 5, sharex=True, sharey=True)
fig.set_size_inches(26, 16, forward=True)

# font = {'weight': 'normal', 'size': 28}
# plt.rc('font', **font)

ax[0, 0].set_xlim(-2, 4.5)
ax[0, 0].set_ylim(-2, 4.5)

fontsize=34
ax[0, 0].set_ylabel('Capto Q\n'+r'$\mathsf{log_{10}\left( c \; [ppm] \right)}$', fontsize=fontsize)
ax[1, 0].set_ylabel('Poros XQ\n'+r'$\mathsf{log_{10}\left( c \; [ppm] \right)}$', fontsize=fontsize)
ax[2, 0].set_ylabel('Poros 50 HQ\n'+r'$\mathsf{log_{10}\left( c \; [ppm] \right)}$', fontsize=fontsize)

for i, loading in enumerate(loadings):
    ax[0, i].set_title(loading, fontsize=fontsize*1.1)
    ax[2, i].set_xlabel(r'$\mathsf{log_{10}\left( c \; [ppm] \right)}$'+'\n'+'PAVIN Feed', fontsize=fontsize)
    for j in range(3):
        ax[j, i].plot([-2, 5], [-2, 5], color='black', linewidth=1)
    
for j, resin in enumerate(resins):
    for i, frac in enumerate(fracs):
        sample = f'{resin}_{frac}'
        ax[j, i].scatter(df_log['pavin_feed'], df_log[sample], s=10)

plt.subplots_adjust(hspace=0.07, wspace=0.07)

plt.savefig('./images/14_frac_correlations_PAVIN_feed.png', dpi=300, bbox_inches='tight')
plt.savefig('./images/14_frac_correlations_PAVIN_feed.pdf', bbox_inches='tight')
```
