# Normalize peak areas to the control

```python
df_areas = hplc.get_areas_df(data, sample_ids)

# Account for dilution 
for i, cont in df_areas.iterrows():
    df_areas.at[i, 'dilution_factor'] = 100/float(cont.frac[-3:])
    for col in ['total', 'large', 'small', 'mab', 'lmw1', 'lmw2', 'lmw3']:
         df_areas.at[i, col] = df_areas.at[i, 'dilution_factor'] * cont[col]
```

```python
df_feed = df_areas[df_areas.frac.str.contains('PAFVIN')]
feed_areas = df_feed.mean(axis=0)
feed_areas = Dict(feed_areas.to_dict())
```

```python
df_norm_all = hplc.get_normalized_area_df(data, sample_ids, feed_areas, df_areas)
df_norm = df_norm_all[3:-1].copy()
hplc.add_volume_midpoints(df_norm, df, cassette=1, start_index=4)
```

# CIP pool

```python
df_cip = df_norm_all[-1:].copy()
df_cip.drop(columns=['frac'], inplace=True)
df_cip /= 41 # approximate concentration factor
df_cip = df_cip.transpose()
df_cip.reset_index(inplace=True)
df_cip.columns = ['species', 'norm_conc']
df_cip['name'] = ['Large agg.', 'Small agg.', 'mAb', 'LMW 1', 'LMW 2', 'LMW 3']
df_cip.drop(index=[4, 5], inplace=True)
```
