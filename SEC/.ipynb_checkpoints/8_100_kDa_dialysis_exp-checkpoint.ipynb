# 100 kDa dialysis experiment

```python
# Load HPLC data
file_path = f'./hplc_data/2022-08-03_100_kDa_dialysis_exp/'
data, sample_ids = hplc.load_hplc_data(file_path, shift_baseline_time=5)

sample_ids.remove('buffer_21194')
```

```python
labels = {
    'Negative_control_21214':'Neg. control',
     'PAFVIN_large_permeate_21198':'Large agg. perm.',
     'PAFVIN_large_retentate_21206':'Large agg. ret.',
     'PAFVIN_small_permeate_21202':'Small agg. perm.',
     'PAFVIN_small_retentate_21210':'Small agg. ret.'
}
```

```python
fig, ax = my_plot.instantiate_fig(xlabel='Time [min]', ylabel='A280 [mAU]')

df = data_pafvin[sample_ids_pafvin[-1]].df
scalar = 0.2
ax.plot(df.time_min, df.uv_280 * scalar, label=f'PAFVIN Feed * {scalar}')

for s in sample_ids:
    ax.plot(data[s].df.time_min, data[s].df.uv_280, label=labels[s])
    
    
my_plot.set_layout(fig, ax, legend=True, fontsize=16, frameon=True)

ax.set_xlim(8, 17.9)
ax.set_ylim(-0.002, 0.02)

plt.savefig('./images/8_100_kDa_dialysis_all.png', dpi=300, bbox_inches='tight')
```
