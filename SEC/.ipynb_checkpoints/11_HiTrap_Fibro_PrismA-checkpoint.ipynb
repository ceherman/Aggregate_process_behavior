# 3 panel plot

```python
fig, ax = plt.subplots(1, 3, sharex=True, sharey=True)
fig.set_size_inches(17, 6, forward=True)
plt.subplots_adjust(wspace=0.02)

fig, ax1, ax2, ax3, lns, labs = hplc.with_sec(df_120, df_norm_120, fig=fig, ax=ax[0], show_twinx=False)
ax[0].set_xlim(0, 23)
ax2.set_ylim(0, 40)
ax3.set_ylim(3, 9)
ax2.set_yticks([])
ax3.set_yticks([])

fig, ax1, ax2, ax3, lns, labs = hplc.with_sec(df_10, df_norm_10, fig=fig, ax=ax[1], show_twinx=False)
ax2.set_ylim(0, 40)
ax3.set_ylim(3, 9)
ax2.set_yticks([])
ax3.set_yticks([])

fig, ax1, ax2, ax3, lns, labs = hplc.with_sec(df_1p5, df_norm_1p5, fig=fig, ax=ax[2], ypad=13, ax3_position=1.28, show_twinx=True)
ax2.set_ylim(0, 40)
ax3.set_ylim(3, 9)

t = ax[0].set_title('120 s', weight='bold')
t = ax[1].set_title('10 s', weight='bold')
t = ax[2].set_title('1.5 s', weight='bold')
for i in range(3):
    xlabel = ax[i].set_xlabel('Volume [ml]')
ylabel = ax[0].set_ylabel('Normalized concentration', labelpad=10)

legend = ax[0].legend(lns, labs, frameon=False, handlelength=1, ncol=2, fontsize=15, columnspacing=0.8, handletextpad=0.3)

# plt.savefig('./images/11_HiTrap_Fibro_PrismA_3_panel.png', dpi=300, bbox_inches='tight')
plt.savefig('./images/11_HiTrap_Fibro_PrismA_3_panel.pdf', bbox_inches='tight')
```
