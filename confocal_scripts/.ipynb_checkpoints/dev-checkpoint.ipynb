## Explanation of power-law colormap normalization ("brightening")

```python
fig, ax = my_plot.instantiate_fig(x=8, xlabel='Normalized fluorescence intensity', ylabel='Probability density')
# ax.set_yscale('log')

data = read_image('./2022-02-19/xz_6_7_1.czi', gamma=1, epsilon=0)
nr = ax.hist(data.flatten(), bins=100, density=True, label='Raw data')

# data = read_image('./2022-02-19/xz_6_7_1.czi', gamma=gamma, epsilon=epsilon)
# nt = ax.hist(data.flatten(), bins=100, density=False, label='Transformed data')

ax.axvline(epsilon, color='black', label='Threshold')

x_vals = np.linspace(0, 1, 1000)
x_vals[x_vals < epsilon] = epsilon
alpha = 1.0 / ((1.0 - epsilon)**gamma)
y_vals = alpha * (x_vals - epsilon)**gamma
ax.plot(x_vals, y_vals*nr[0].max(), '--', color='black', label='Power law (scaled)')

my_plot.set_layout(fig, ax, legend=True)
fig.savefig(f'./{image_folder}/intensity_scaling.png', dpi=300)
```

```python
fig, ax = my_plot.instantiate_fig(x=8, xlabel='Normalized fluorescence intensity', ylabel='Probability density')
ax.set_ylim(0, 2)

data = read_image('./2022-02-19/xz_6_7_1.czi', gamma=1, epsilon=0)
nr = ax.hist(data.flatten(), bins=100, density=True, label='Raw data', alpha=0.7)

data = read_image('./2022-02-19/xz_6_7_1.czi', gamma=gamma, epsilon=epsilon)
nt = ax.hist(data.flatten(), bins=100, density=True, label='Transformed data', alpha=0.7)

ax.axvline(epsilon, color='black', label='Threshold')

x_vals = np.linspace(0, 1, 1000)
x_vals[x_vals < epsilon] = epsilon
alpha = 1.0 / ((1.0 - epsilon)**gamma)
y_vals = alpha * (x_vals - epsilon)**gamma
ax.plot(x_vals, y_vals, '--', color='black', label='Power law')

my_plot.set_layout(fig, ax, legend=True)
fig.savefig(f'./{image_folder}/intensity_scaling_with_transformed.png', dpi=300)
```
